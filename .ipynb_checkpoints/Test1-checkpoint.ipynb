```python
#import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore')
```

```python
df_train=pd.read_csv("train.csv")
df_test=pd.read_csv("test.csv")
df_sample=pd.read_csv("sample_submission.csv")
combined=pd.concat([df_train,df_test])
combined['Dependents'].loc[combined['Dependents'] == '3+'] = 3
```

```python
df_train.shape
```

## Filling NaN values with the mode

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=4, weights="uniform")

cols=["Dependents","Gender","Loan_Amount_Term","Married","Self_Employed","Credit_History"]
combined[cols]=combined[cols].fillna(combined.mode().iloc[0])
combined[["CoapplicantIncome","Loan_Amount_Term","LoanAmount"]] = imputer.fit_transform(combined[["CoapplicantIncome","Loan_Amount_Term","LoanAmount"]])
#combined[["Credit_History"]] = imputer.fit_transform(combined[["Credit_History"]])
```

```python
combined['CoapplicantIncome'].loc[combined['CoapplicantIncome'] == 0] = 0
combined['CoapplicantIncome'].loc[combined['CoapplicantIncome'] != 0] = 1
```

## Finalizing dataset

```python
combined = pd.get_dummies(combined,columns=["Education","Gender","Married","Property_Area"])
combined[["Credit_History","Dependents"]]=combined[["Credit_History","Dependents"]].astype(int)
combined.drop(["CoapplicantIncome","Gender_Female","Self_Employed","Married_No","Property_Area_Urban","Education_Graduate","Property_Area_Semiurban"],axis=1,inplace=True)
```

```python
sns.distplot(combined["ApplicantIncome"])
```

## Feature Scaling 1) robust scaler


```python
from sklearn.preprocessing import normalize
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
combined[['ApplicantIncome','LoanAmount','Loan_Amount_Term']] = scaler.fit_transform(combined[['ApplicantIncome', 'LoanAmount','Loan_Amount_Term']])
```

## Model Training

```python
df_train=combined.iloc[:614,]
df_test=combined.iloc[614:,]
```

```python
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()

df_train["Loan_Status"]=le.fit_transform(df_train.Loan_Status)
df_train.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
X=df_train.drop(["Loan_Status"],axis=1)
y=df_train["Loan_Status"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import f1_score
```

```python
 
rf = XGBClassifier(n_estimators=200,learning_rate=0.1)
rf.fit(X_train,Y_train)
print("Dropping column")
predictions = rf.predict(X_test)
print("F1 score " + f1_score(Y_test, predictions).astype(str))
score=round(rf.score(X_test, Y_test) * 100, 2)
print("Accuracy Score = "+score.astype(str))
```

```python
 
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
print("Dropping column")
predictions = rf.predict(X_test)
print("F1 score " + f1_score(Y_test, predictions).astype(str))
score=round(rf.score(X_test, Y_test) * 100, 2)
print("Accuracy Score = "+score.astype(str))
```
